```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

features = df[['평균 1~5단계 중 1단계 이상 완료 도서 수','평균 학습 단어 수(3단계 완료한 도서의 단어 수 합)','평균 월간 학습 일수','평균 발음점수','1단계 학습 시간(초)','2단계 학습 시간','3단계 학습 시간','4단계 학습 시간','5단계 학습 시간']] # 분석에 필요한 특성 선택

# from sklearn.preprocessing import RobustScaler
# scaler = RobustScaler()
# scaled_features = scaler.fit_transform(features)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)

#데이터 표준화
# scaler = StandardScaler()
# scaled_features = scaler.fit_transform(features)
```

```python
# 2차원 그래프 그리기
plt.figure(figsize=(8, 6))  # 그래프의 크기를 조절 (선택 사항)

plt.scatter(scaled_features[:, 0], scaled_features[:, 1], alpha=0.5)  # 산점도 그리기

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Dimensionality Reduction')

plt.grid()  # 그리드 추가 (선택 사항)

plt.show()
```

```python
from sklearn.decomposition import PCA

# PCA를 사용한 차원 축소
pca = PCA(n_components=2)  # 2차원으로 축소
reduced_features = pca.fit_transform(scaled_features)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA를 사용한 차원 축소
pca = PCA(n_components=2)  # 2차원으로 축소
reduced_features = pca.fit_transform(scaled_features)

# 2차원 그래프 그리기
plt.figure(figsize=(8, 6))  # 그래프의 크기를 조절 (선택 사항)

plt.scatter(reduced_features[:, 0], reduced_features[:, 1], alpha=0.5)  # 산점도 그리기

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Dimensionality Reduction')

plt.grid()  # 그리드 추가 (선택 사항)

plt.show()
```

```python
from sklearn.metrics import silhouette_score

# 군집 개수 범위 설정
min_clusters = 3
max_clusters = 10

# 군집 개수별 실루엣 계수 저장할 리스트
silhouette_scores = []

# 군집 수별로 K-Means 모델 생성 및 실루엣 계수 계산
for num_clusters in range(min_clusters, max_clusters + 1):
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(scaled_features)

    # 실루엣 계수 계산
    silhouette_avg = silhouette_score(reduced_features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# 실루엣 계수 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score by Number of Clusters')
plt.show()
```

```python
# 군집 개수 설정
num_clusters = 4

# K-Means 모델 생성
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(scaled_features)

# 실루엣 계수 계산
silhouette_avg = silhouette_score(reduced_features, cluster_labels)
print(f"실루엣 계수 (군집 개수 {num_clusters}): {silhouette_avg}")

# 군집 결과 확인
df['cluster'] = cluster_labels

# 군집별 데이터 개수 출력
cluster_counts = df['cluster'].value_counts()
print("군집별 데이터 개수:")
print(cluster_counts)
```

```python
import numpy as np
import matplotlib.cm as cm
from sklearn.metrics import silhouette_samples, silhouette_score
from matplotlib import font_manager, rc


# 각 데이터 포인트의 실루엣 계수를 계산하고 'silhouette_values' 변수에 저장
silhouette_values = silhouette_samples(reduced_features, cluster_labels)

# 각 데이터 포인트의 실루엣 계수를 데이터프레임에 추가
data_with_silhouette = pd.DataFrame(data=df, columns=['평균 1~5단계 중 1단계 이상 완료 도서 수', '평균 학습 단어 수(3단계 완료한 도서의 단어 수 합)', '평균 월간 학습 일수', '평균 발음점수', '1단계 학습 시간(초)', '2단계 학습 시간', '3단계 학습 시간', '4단계 학습 시간', '5단계 학습 시간'])
data_with_silhouette["Cluster"] = cluster_labels
data_with_silhouette["Silhouette"] = silhouette_values

# 각 클러스터의 실루엣 계수 계산
cluster_avg_silhouette_scores = []
for i in range(num_clusters):
    cluster_i_silhouette_values = data_with_silhouette[data_with_silhouette["Cluster"] == i]["Silhouette"]
    avg_silhouette_i = np.mean(cluster_i_silhouette_values)
    cluster_avg_silhouette_scores.append(avg_silhouette_i)

# 각 클러스터의 실루엣 계수 출력
for i, score in enumerate(cluster_avg_silhouette_scores):
    print(f"Cluster {i}: Average Silhouette Score = {score}")

# 전체 데이터의 실루엣 계수 평균 출력
avg_silhouette = silhouette_score(reduced_features, cluster_labels)
print(f"Overall Average Silhouette Score = {avg_silhouette}")

# 실루엣 계수 시각화
def visualize_silhouette(cluster_labels, silhouette_values):
    fig, ax = plt.subplots(1, 1)
    fig.set_size_inches(18, 7)

    y_lower = 10
    for i in range(num_clusters):
        ith_cluster_silhouette_values = silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / num_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    ax.set_title("Silhouette plot for various clusters.")
    ax.set_xlabel("Silhouette coefficient values")
    ax.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax.axvline(x=avg_silhouette, color="red", linestyle="--", linewidth=4)

    ax.set_yticks([])  # Clear the yaxis labels / ticks
    ax.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    plt.show()

# 실루엣 계수 시각화 호출
visualize_silhouette(cluster_labels, silhouette_values)
from sklearn.cluster import KMeans


```
